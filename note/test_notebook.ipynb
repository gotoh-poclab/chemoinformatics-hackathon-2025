## 記述子の取得

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, PandasTools
import pandas as pd

# Mol列を追加（分子オブジェクト）
PandasTools.AddMoleculeColumnToFrame(df, smilesCol='SMILES', molCol='Molecule')

df.head()
```

```python
from rdkit.Chem import Descriptors

# 使用したい記述子の関数リスト（必要ならここで追加・削除）
descriptor_functions = {
    "MolWt": Descriptors.MolWt,
    "MolLogP": Descriptors.MolLogP,
    "TPSA": Descriptors.TPSA,
    "NumHDonors": Descriptors.NumHDonors,
    "NumHAcceptors": Descriptors.NumHAcceptors,
    "NumRotatableBonds": Descriptors.NumRotatableBonds,
    "FractionCSP3": Descriptors.FractionCSP3,
    "RingCount": Descriptors.RingCount,
    "HeavyAtomCount": Descriptors.HeavyAtomCount,
}

from rdkit import Chem
from rdkit.Chem import PandasTools

def add_descriptors(df, smiles_col="SMILES", mol_col="Molecule"):
    # 分子構造を追加（既にある場合は skip 可能）
    if mol_col not in df.columns:
        PandasTools.AddMoleculeColumnToFrame(df, smilesCol=smiles_col, molCol=mol_col)

    # 各記述子を計算して列として追加
    for name, func in descriptor_functions.items():
        df[name] = df[mol_col].apply(lambda mol: func(mol) if mol is not None else None)

    return df
```

```python
df_train = add_descriptors(df)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np

# 特徴量とラベルの準備（すでに df_desc に記述子がある前提）
feature_cols = [
    "MolWt", "MolLogP", "TPSA", "NumHDonors", "NumHAcceptors",
    "NumRotatableBonds", "FractionCSP3", "RingCount", "HeavyAtomCount"
]
df_model = df.dropna(subset=feature_cols + ["HIT"]).copy()
X = df_model[feature_cols].astype(float).values
y = LabelEncoder().fit_transform(df_model["HIT"])

# 5-fold交差検証
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 各foldのconfusion matrixを出力
for fold_idx, (train_idx, val_idx) in enumerate(cv.split(X, y), start=1):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    model = RandomForestClassifier(class_weight="balanced", random_state=42)
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)

    cm_train = confusion_matrix(y_train, y_train_pred)
    cm_val = confusion_matrix(y_val, y_val_pred)

    # 可視化
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    disp1 = ConfusionMatrixDisplay(cm_train, display_labels=["Inactive", "Active"])
    disp2 = ConfusionMatrixDisplay(cm_val, display_labels=["Inactive", "Active"])

    disp1.plot(ax=axes[0], cmap="Blues", values_format="d")
    axes[0].set_title(f"Fold {fold_idx} - Training")

    disp2.plot(ax=axes[1], cmap="Oranges", values_format="d")
    axes[1].set_title(f"Fold {fold_idx} - Validation")

    plt.suptitle(f"Confusion Matrices - Fold {fold_idx}", fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import matthews_corrcoef

# MCCの記録用リスト
mcc_train_scores = []
mcc_val_scores = []

# 再度、foldごとにループ（混同行列と同時に MCC を出す）
for fold_idx, (train_idx, val_idx) in enumerate(cv.split(X, y), start=1):
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    model = RandomForestClassifier(class_weight="balanced", random_state=42)
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)

    # MCCの計算
    mcc_train = matthews_corrcoef(y_train, y_train_pred)
    mcc_val = matthews_corrcoef(y_val, y_val_pred)

    mcc_train_scores.append(mcc_train)
    mcc_val_scores.append(mcc_val)

    print(f"Fold {fold_idx}:")
    print(f"  Training MCC:   {mcc_train:.3f}")
    print(f"  Validation MCC: {mcc_val:.3f}")
    print()

# 全体平均
print("Average Training MCC:   {:.3f} ± {:.3f}".format(
    np.mean(mcc_train_scores), np.std(mcc_train_scores)
))
print("Average Validation MCC: {:.3f} ± {:.3f}".format(
    np.mean(mcc_val_scores), np.std(mcc_val_scores)
))
```

## テストデータの予測

```python
# 例: test_public.tsv を読み込み
X_test = df_test_public[feature_cols].astype(float)
```

```python
# 予測（0=Inactive, 1=Active）
y_pred = clf_final.predict(X_test)
y_proba = clf_final.predict_proba(X_test)[:, 1]  # Active の確率

# 結果を保存
df_test_public["Predicted_Label"] = y_pred
df_test_public["Predicted_Prob_Active"] = y_proba
```

```python
# 予測ラベルを 0/1 → Active/Inactive に変換
df_test_public["prediction"] = df_test_public["Predicted_Label"].map({1: "Active", 0: "Inactive"})

df_test_public
```

```python
# 必要ライブラリ
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, matthews_corrcoef
import matplotlib.pyplot as plt


# Active/Inactive を数値に変換
label_map = {"Inactive": 0, "Active": 1}
y_true = df_test_public["HIT"].map(label_map).values
y_pred = df_test_public["prediction"].map(label_map).values

# 混同行列と MCC の計算
cm = confusion_matrix(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

# 混同行列を可視化
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Inactive", "Active"])
disp.plot(cmap="Blues", values_format="d")
plt.title(f"Confusion Matrix (Test Public) - MCC = {mcc:.3f}")
plt.show()

# MCC を出力
print(f"MCC: {mcc:.3f}")
```
